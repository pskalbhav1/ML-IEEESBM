```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

x=np.linspace(20,40,100)
y1=[((-3)*i + 80) for i in x if i<=30]
y2=[((2*i)-20) for i in x if i>30]
y3=[(2*i) for i in x if i<=30]
y4=[((-3*i)+110) for i in x if i>30]

y5=y1+y2
y6=y3+y4
plt.plot(x,y5,'.',color='b',label='line1-dotted')
plt.plot(x,y6,'--', color='r',label='line2-dashed')

plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.legend()
plt.title("Plot with two or more lines with different styles")
plt.show()
```
